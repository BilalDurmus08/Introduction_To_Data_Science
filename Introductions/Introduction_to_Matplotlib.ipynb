## which one should we use? (pyplot vs matplotlib OO method)
* When plotting something quickly, okey to use pyplot method
* When plotting something more advanced, use the OO method 

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50
```

```python
## Plot Method
over_50.plot(kind="scatter",
             x="age",
             y="chol", 
             c="target");
```

```python
## OO Method mixed with pyplot method
fig ,ax = plt.subplots(figsize=(10,6))
over_50.plot(kind = "scatter",
            x = "age",
            y = "chol", 
            c = "target",
            ax = ax);
# ax.set_xlim([45, 65])  ## set a limit
```

```python
## OO method mixed with pyplot method
fig, ax = plt.subplots(figsize=(10,6))

# Plot the data
scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"])

# Customize the plot
ax.set(title="Heart Diesase and Cholestrol levels.",
      xlabel="Age",
      ylabel="Chol")

# Add legend method
ax.legend(*scatter.legend_elements(), title="Target");

# Add mean
ax.axhline(y=over_50["chol"].mean(), linestyle="--")
```

```python
# Creating fig with contains 2 axes
fig, (ax0, ax1) = plt.subplots(figsize=(10,10),
                               ncols=1,
                               nrows=2,
                               sharex=True)

# Plot with some data
scatter = ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"],
                     cmap="winter") #Customized the color of dots!

# Customize the data
ax0.set(title="Age-Chol Plot table",
        ylabel="Chol")

# Adding legend 
ax0.legend(*scatter.legend_elements(), title="Targets")

# Adding mean
ax0.axhline(over_50["chol"].mean())

#Setting limitations
ax0.set_ylim(150,450)



## Second Axis Starting process
# plot with some data
scatter = ax1.scatter(x=over_50['age'],
                      y=over_50['thalach'],
                      c=over_50['target'],
                      cmap="summer")
# Customize the plot
ax1.set(title="Age-Thalach Plot Table",
        xlabel="Age",
        ylabel="Thalach")

#Giving limitations
ax1.set_xlim(50,75)
ax1.set_ylim(80,200)

# Add legend line
ax1.legend(*scatter.legend_elements(), title="Target")

# Add mean
ax1.axhline(y=over_50['thalach'].mean(), linestyle="--");

## Adding title to the figure
fig.suptitle('Heart diesease Figure', fontsize=16, fontweight="bold");
```
